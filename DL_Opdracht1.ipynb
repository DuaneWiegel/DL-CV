## 1 Exploratieve Data Analyse

### Dataset inladen

```python
# Data Sciene Libraries importeren 
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Dataset laden 
df = pd.read_csv('train.csv')

# dataset weergeven 
display(df.head())
```

```python
# basis info over dataset
print(df.info())
print(df.describe())
print(df.isnull().sum())

# unieke waardes controleren voor prijs 
print(df['Price'].unique())

# de frequentie van elke unieke waarde in de price-kolom bekijken 
print(df['Price'].value_counts())

# Formatter voor nette as-notatie
import matplotlib.ticker as mtick

# Histogram plot aanmaken om de verdeling van de Price-kollom te visualiseren
plt.figure(figsize=(10, 5))
 
sns.histplot(df["Price"], bins=30, kde=True)
plt.title("Verdeling van Price")
plt.xlabel("Price")
plt.ylabel("Frequentie")

# Pak huidige as op om ticks/formattering aan te passen
ax = plt.gca()

# Zet x-as ticks op stappen van 100.000
ax.xaxis.set_major_locator(mtick.MultipleLocator(100000))

# Toon volledige getallen met duizendtallen-scheiding (bijv. 100,000)
ax.xaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))

# Draai x-labels zodat ze niet overlappen
plt.xticks(rotation=90)

# Zorg dat alles netjes binnen het figuur past
plt.tight_layout()

# Toon de plot
plt.show()

# ...existing code...
```
